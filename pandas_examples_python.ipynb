```python
# prompt: find out basic stats information on my dataframe df, calculate measurements such as mean, median, mode, variance, std, range, skewness and kurtosis

# Import necessary libraries if not already imported
import pandas as pd

# Calculate range
range_salary = df['salary'].max() - df['salary'].min()

# Calculate skewness
skewness_salary = df['salary'].skew()

# Calculate kurtosis
kurtosis_salary = df['salary'].kurt()

# Print the results
print(f"Mean: {df['salary'].mean()}")
print(f"Median: {df['salary'].median()}")
print(f"Mode: {df['salary'].mode()[0]}")  # Access the first mode value
print(f"Variance: {df['salary'].var()}")
print(f"Standard Deviation: {df['salary'].std()}")
print(f"Range: {range_salary}")
print(f"Skewness: {skewness_salary}")
print(f"Kurtosis: {kurtosis_salary}")
```

```python
# prompt: create distplot for all the the columns in dataframe df1

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df1 is already defined and contains the data
# If not, load your DataFrame here:
# df1 = pd.read_csv('/content/Universities.csv')

for col in df1.columns:
  if pd.api.types.is_numeric_dtype(df1[col]): # Check if the column is numeric
    sns.distplot(df1[col])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# prompt: create a boxplot on accept column in dataframe df1

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 is already defined and contains the 'Accept' column
# If not, load your DataFrame here:
# df1 = pd.read_csv('/content/Universities.csv')

sns.boxplot(y=df1['Accept']) # Create the boxplot
plt.xlabel('Accept') # Add x-axis label
plt.ylabel('Values') # Add y-axis label
plt.title('Boxplot of Accept') # Add title
plt.show() # Show the plot
```
